```python
import pandas as pd
import numpy as np

concrete_data = pd.read_csv('https://cocl.us/concrete_data')
concrete_data.head()
```

```python
concrete_data_columns = concrete_data.columns
```

```python
predictors = concrete_data[concrete_data_columns[concrete_data_columns != 'Strength']]
target = concrete_data['Strength']
```

```python
n_cols = predictors.shape[1]
```

```python
import keras
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
def regression_model():

    model = Sequential()
    model.add(Dense(10, activation='relu', input_shape=(n_cols,)))
    model.add(Dense(1))
    
  
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model
```

```python
model = regression_model()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(predictors, target, test_size=0.30)
model.fit(predictors, target, validation_split=0.3, epochs=50)
```

```python
from sklearn.metrics import mean_squared_error
y_hat = model.predict(X_test)
```

```python
#The Mean Squared Error is:
mean_squared_error(y_test, y_hat)
```

```python
from itertools import repeat

MSEs = []
for i in repeat(None,50):
    i = X_train, X_test, y_train, y_test
    X_train, X_test, y_train, y_test = train_test_split(predictors, target, test_size=0.30)
    model.fit(predictors, target, validation_split=0.3, epochs=50)
    y_hat = model.predict(X_test)
    m = mean_squared_error(y_test, y_hat)
    MSEs.append(m)
```

```python
# List of 50 Mean Squared Errors: 
print("These are the 50 Mean Squared Errors: ", MSEs)
print("The mean is: ", np.mean(MSEs))
print("The Standard Deviation is: ", np.std(MSEs))
```

```python
#Normalize the data
predictors_norm = (predictors - predictors.mean()) / predictors.std()
predictors_norm.head()
```

```python
#Getting the Mean Squared Errors with the Normalized data.

X_train, X_test, y_train, y_test = train_test_split(predictors_norm, target, test_size=0.30)
model.fit(predictors_norm, target, validation_split=0.3, epochs=50)
y_hat = model.predict(X_test)
mean_squared_error(y_test, y_hat)

MSEs_norm = []
for i in repeat(None,50):
    i = X_train, X_test, y_train, y_test
    X_train, X_test, y_train, y_test = train_test_split(predictors_norm, target, test_size=0.30)
    model.fit(predictors_norm, target, validation_split=0.3, epochs=50)
    y_hat = model.predict(X_test)
    m = mean_squared_error(y_test, y_hat)
    MSEs_norm.append(m)
```

```python
# List of 50 Mean Squared Errors with the Normalized Data: 
print("These are the 50 Mean Squared Errors: ", MSEs_norm)
print("The mean is: ", np.mean(MSEs_norm))
print("The Standard Deviation is: ", np.std(MSEs_norm))
```

```python
print("The initial mean is: ", np.mean(MSEs))
print("The mean of the normalized data is: ", np.mean(MSEs_norm))
print("The mean of the initial data is {} times larger than the normalized mean".format((np.mean(MSEs)/(np.mean(MSEs_norm)))))
```

```python
#Getting the Mean Squared Errors with 100 epochs.

X_train, X_test, y_train, y_test = train_test_split(predictors_norm, target, test_size=0.30)
model.fit(predictors_norm, target, validation_split=0.3, epochs=100)
y_hat = model.predict(X_test)
mean_squared_error(y_test, y_hat)

MSEs_norm_100 = []
for i in repeat(None,50):
    i = X_train, X_test, y_train, y_test
    X_train, X_test, y_train, y_test = train_test_split(predictors_norm, target, test_size=0.30)
    model.fit(predictors_norm, target, validation_split=0.3, epochs=100)
    y_hat = model.predict(X_test)
    m = mean_squared_error(y_test, y_hat)
    MSEs_norm_100.append(m)
```

```python
# List of 50 Mean Squared Errors with 100 epochs: 
print("These are the 50 Mean Squared Errors: ", MSEs_norm_100)
print("The mean is: ", np.mean(MSEs_norm_100))
print("The Standard Deviation is: ", np.std(MSEs_norm_100))
```

```python
print("The Mean Squared Error with 50 Epochs is {} times larger than with 100 Epochs".format((np.mean(MSEs_norm)/(np.mean(MSEs_norm_100)))))
```

```python
# Now Let's create the new model with 3 hidden layers
def regression_model2():

    model2 = Sequential()
    model2.add(Dense(10, activation='relu', input_shape=(n_cols,)))
    model2.add(Dense(10, activation='relu'))
    model2.add(Dense(10, activation='relu'))
    model2.add(Dense(1))
    
    model2.compile(optimizer='adam', loss='mean_squared_error')
    return model2

model2 = regression_model2()

X_train, X_test, y_train, y_test = train_test_split(predictors_norm, target, test_size=0.30)
model2.fit(predictors_norm, target, validation_split=0.3, epochs=50)
y_hat = model2.predict(X_test)
mean_squared_error(y_test, y_hat)

#Let's repeat the process 50 times                     
                     
MSEs_norm2 = []
for i in repeat(None,50):
    i = X_train, X_test, y_train, y_test
    X_train, X_test, y_train, y_test = train_test_split(predictors_norm, target, test_size=0.30)
    model.fit(predictors_norm, target, validation_split=0.3, epochs=50)
    y_hat = model2.predict(X_test)
    m = mean_squared_error(y_test, y_hat)
    MSEs_norm2.append(m)
```

```python
# List of 50 Mean Squared Errors with 3 layers: 
print("These are the 50 Mean Squared Errors: ", MSEs_norm2)
print("The mean is: ", np.mean(MSEs_norm2))
print('The mean of the model with 3 layers is {} times larger than with 1 layer'.format(np.mean(MSEs_norm2)/np.mean(MSEs_norm)))
```
